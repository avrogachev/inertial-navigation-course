```python
def CNFparser(filename):
    '''
    Тут мы расписываем парсер файла данных - при необходимости, строим дополнительные колонки и 
    делаем прочие манипуляции
    '''
    CNFnames = ['Time','AccZ','AccX','AccY','OmZ','OmX','OmY','H']
    CNF = pd.read_csv(filename, sep = '\s+' ,header = None, names = CNFnames, skiprows = 1)
    # добавим колонку с некоторым инвариантом, удобным для поиска интервала выставки:
    CNF['Invariant'] = CNF.AccX*CNF.OmX +  CNF.AccY*CNF.OmY +  CNF.AccZ*CNF.OmZ 
    return CNF
```

Изучив графики, я решил осуществить выставку на интервале с 1940 до 2020 секунды (12000 - 45000 измерения), чтобы определить начальную ориентацию БИНС

```python
def algebraic_alignment_naive(A, phi, h, a, e2, u):
    # выставка во втором варианте из методички, причём с известной точной широтой места проведения выставки
    # я её имплементировал, но решил пользоваться первым вариантом - это следующая ячейка
    L = np.zeros((3, 3))
    init = 12000 
    end = 45000  # получается 80 секунд
    om_x = CNF['OmX'][:].mean()
    om_y = CNF['OmY'][:].mean()  
    om_z = CNF['OmZ'][:].mean()
    f_x = CNF['AccX'][:].mean()
    f_y = CNF['AccY'][:].mean()
    f_z = CNF['AccZ'][:].mean()
    
    theta = math.atan2(f_x, math.sqrt(f_y*f_y+f_z*f_z))
    gamm = -math.atan2(f_z, f_y)
    
    
    means = np.mean(A[init:end, 1:], axis = 0).reshape(6,1)
    theta = math.atan2(means[0], math.sqrt(means[1]**2+means[2]**2))
    gamm = -math.atan2(means[2], means[1])
    
    L[:, 2:3] = means[0:3] / 9.812357#(g_phi(phi, h, a, e2))
    L[:, 1:2] = (means[3:6] - (L[:, 2:3] * u * sin(phi))) / (u * cos(phi))
    L[:, 0:1] = np.cross(L[:, 1:2].T, L[:, 2:3].T).T
    L[:, 2:3] /= np.linalg.norm(L[:, 2:3], 2)
    L[:, 1:2] /= np.linalg.norm(L[:, 1:2], 2)
    L[:, 0:1] /= np.linalg.norm(L[:, 0:1], 2)
    psi = math.atan2(L[0][0], L[0][1])
    return L, theta, gamm, psi, means
```

```python
def initial_algebraic_alignment():
    '''
    Это функция начальной алгебраической выставки - в глобальные переменные записываются значения углов
    '''
    global g
    global L
    global theta, gamm, phi, psi
    
    init = 12000
    end = 45000  # 15 секунд
    om_x = CNF['OmX'][init:end].mean()
    om_y = CNF['OmY'][init:end].mean()  
    om_z = CNF['OmZ'][init:end].mean()
    f_x = CNF['AccX'][init:end].mean()
    f_y = CNF['AccY'][init:end].mean()
    f_z = CNF['AccZ'][init:end].mean()
    
    theta = math.atan2(f_x, math.sqrt(f_y**2+f_z**2))
    gamm = -math.atan2(f_z, f_y)
    om = math.sqrt(om_x**2+om_y**2+om_z**2)
    
    phi = math.atan2(om_z, math.sqrt(om_x**2+om_y**2))
    g_mean = math.sqrt(f_x**2+f_y**2+f_z**2)
    # третий столбец по показаниям акселерометров
    l13 = sin(theta)
    l23 = cos(theta)*cos(gamm)
    l33 = -cos(theta)*sin(gamm)
    # второй и первый по показаниям ДУС
    l12 = (om_x - u*l13*sin(phi))/(u*cos(phi))
    l22 = (om_y - u*l23*sin(phi))/(u*cos(phi))
    l32 = (om_z - u*l33*sin(phi))/(u*cos(phi))
    norm_coeff = sqrt(l12**2+l22**2+l32**2)
    l12 = l12/norm_coeff #нормировка
    l22 = l22/norm_coeff
    l32 = l32/norm_coeff
    
    #теперь нужно векторно умножить второй на третий
    l11 = l22*l33 - l32*l23
    l21 =-l12*l33 + l13*l32
    l31 = l12*l23 - l22*l13
    
    L = np.matrix([[l11, l12, l13], [l21, l22, l23], [l31, l32, l33]])   #записали в матрицу

    psi = math.atan2(l11,l12)

    theta = math.asin(L.item(0,2))
    
    gamm = -math.atan2(L.item(2,2),L.item(1,2))
    
    # и по полученным углам перевычислили матрицу ориентации:
    L = np.array([[sin(psi)*cos(theta)                      , cos(psi)*cos(theta)                                , sin(theta)],
         [cos(psi)*sin(gamm)-sin(psi)*sin(theta)*cos(gamm), -sin(psi)*sin(gamm)-cos(psi)*sin(theta)*cos(gamm), cos(theta)*cos(gamm)],
         [cos(psi)*cos(gamm)+sin(psi)*sin(theta)*sin(gamm), -sin(psi)*cos(gamm)+cos(psi)*sin(theta)*sin(gamm), -cos(theta)*sin(gamm)]])
    L = L.reshape((3,3))
```
